<h1>Filtering Data</h1>

```python
df['region_txt'].unique()
```

```python
df['attacktype1_txt'].unique()
```

```python
filter1 = df[df['region_txt']== 'Middle East & North Africa']
```

```python
filter2 = filter1[filter1['attacktype1_txt'] == 'Bombing/Explosion']
```

```python
filter3 = filter2[filter2['nkill'] >= 10]
```

```python
filter4 = filter3[filter3['iyear'] >= 2000]
```

<h1>Data Aggregation</h1>

```python
ME_terror = filter4.groupby(['longitude','latitude']).size().reset_index(name= 'incident_count')
```

```python
ME_terror
```

<H1>Visualization of GeoSpatial using FOLIUM</H1>

```python
import folium
from folium.plugins import HeatMap
```

<h1>Regional Analysis</h1>

```python
heat_df = ME_terror[['longitude','latitude', 'incident_count']].values.tolist()
```

```python
my_map = folium.Map(location = [30,40], zoom_start = 4)
```

```python
HeatMap(heat_df).add_to(my_map)
```

```python
my_map
```

```python
my_map.save('terrorism_middleeast.html')
```

<h1>Regional Analysis</h1>

```python
regional = df[df['region_txt'] == 'Southeast Asia']
regional1 = regional[regional['attacktype1_txt'] == 'Armed Assault']
regional2 = regional1[regional1['nkill'] < 1]
regional3 = regional2[regional2['iyear'] < 2000]

regional_agg = regional3.groupby(['latitude','longitude']).size().reset_index(name= 'incident_count')
heat_regional = regional_agg[['latitude','longitude', 'incident_count']].values.tolist()
regional_map = folium.Map(location = [12,122], zoom_start = 4)
HeatMap(heat_regional).add_to(regional_map)
my_map.save('southeast_asia.html')
regional_map
```

<h1>Crime/Attack Analysis</h1>

```python
crime = df[df['attacktype1_txt'] == 'Facility/Infrastructure Attack']
crime1 = crime[crime['region_txt'] == 'Western Europe']
crime2 = crime1[crime1['nkill'] > 1]

crime_agg = crime2.groupby(['latitude','longitude']).size().reset_index(name= 'incident_count')
heat_crime = crime_agg[['latitude','longitude', 'incident_count']].values.tolist()
crime_map = folium.Map(location = [45,10], zoom_start = 4)
HeatMap(heat_crime).add_to(crime_map)
crime_map.save('western_europe.html')
crime_map
```

<h1>Nationality Analysis</h1>

```python
nationality = df[df['region_txt'].str.contains('North America')]
nationality1 = nationality[nationality['nkill'] < 10]
nationality2 = nationality1[nationality1['natlty1_txt'] == 'United States']

nationality_agg = nationality1.groupby(['latitude','longitude']).size().reset_index(name= 'incident_count')
heat_nationality = nationality_agg[['latitude','longitude', 'incident_count']].values.tolist()
nationality_map = folium.Map(location = [30, -96], zoom_start = 4)
HeatMap(heat_nationality).add_to(nationality_map)
nationality_map.save('United_States.html')
nationality_map
```

<h1>Casualty Analysis</h1>

```python
casualty = df[df['region_txt'].str.contains('East Asia')]
casualty1 = casualty[casualty['nkill'] == 1]

casualty_agg = casualty1.groupby(['latitude','longitude']).size().reset_index(name= 'incident_count')
heat_casualty = casualty_agg[['latitude','longitude', 'incident_count']].values.tolist()
casualty_map = folium.Map(location = [30, 110], zoom_start = 4)
HeatMap(heat_casualty).add_to(casualty_map)
casualty_map.save('East_Asia.html')
casualty_map
```
